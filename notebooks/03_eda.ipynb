Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

Load Cleaned Dataset

```python
df = pd.read_csv("../data/cleaned_ecommerce.csv")
```

```python
df.shape
```

Event Type Distribution

```python
df['event_type'].value_counts()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='event_type', data=df)
plt.title("Distribution of Event Types")
plt.savefig("../images/event_type_distribution.png")
plt.show()
```

Price Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df['price'], bins=50)
plt.title("Product Price Distribution")
plt.savefig("../images/product_price_distribution.png")
plt.show()
```

Top 10 Brands

```python
top_brands = df['brand'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_brands.index, y=top_brands.values)
plt.xticks(rotation=45)
plt.title("Top 10 Brands")
plt.savefig("../images/top_products.png")
plt.show()
```

Hourly Activity of Users

```python
plt.figure(figsize=(10,5))
sns.countplot(x='hour', data=df)
plt.title("User Activity by Hour")
plt.show()
```

Monthly Activity

```python
plt.figure(figsize=(6,4))
sns.countplot(x='month', data=df)
plt.title("Monthly Activity")
plt.show()
```

Top Categories

```python
top_categories = df['category_code'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_categories.values, y=top_categories.index)
plt.title("Top Product Categories")
plt.show()
```

Average Price per Event Type

```python
df.groupby('event_type')['price'].mean()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='event_type', y='price', data=df)
plt.title("Price Distribution by Event Type")
plt.show()
```

Correlation Matrix

```python

numeric_df = df.select_dtypes(include=['int64','float64'])

plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```
